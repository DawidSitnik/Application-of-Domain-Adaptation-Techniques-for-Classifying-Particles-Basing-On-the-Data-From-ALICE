```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

model_name = 'source'

plt.figure(figsize=(15, 10))
plt.title("Adaptation Quality Of Clusterts", fontsize=20)
plt.xlabel("Cluster", fontsize=15)
plt.ylabel("Adaptation Quality", fontsize=15)


for particle_name in ['electrons', 'protons', 'pions', 'kaons']:
    df = pd.read_csv(f'./{model_name}_{particle_name}_clusters_summaries.csv')
    df = df.loc[df['a-distance'] != 0]
#     x1 = np.linspace(0, 1, len(df))
    df = df.sort_values(by='a-distance').reset_index(drop=True)
    plt.plot(df['a-distance'], label=particle_name)
#     plt.savefig('./a.png')
    
plt.legend(loc=2, prop={'size': 15})
```
